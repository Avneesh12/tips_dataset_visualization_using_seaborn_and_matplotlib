```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tip_df = sns.load_dataset('tips')
tip_df.head()
```

```python
# Line plot for totalbill and tip 
plt.figure(figsize=(12,6))
sns.lineplot(x='total_bill',y='tip',data=tip_df,marker='o',color='y')
plt.show()
```

```python
# Line plot for totalbill and tip for male and female
plt.figure(figsize=(12,6))
sns.lineplot(x='total_bill',y='tip',data=tip_df,marker='o',color='y',hue='sex')
```

```python
# Line plot for totalbill and tip for smoker
plt.figure(figsize=(12,6))
sns.lineplot(x='total_bill',y='tip',data=tip_df,marker='o',color='y',hue='smoker')
```

```python
# Bar plot for total bill on days basis on the sex
sns.barplot(x='day',y='total_bill',data=tip_df,hue='sex')
```

```python
# Bar plot for tip on days basis on the sex
sns.barplot(x='day',y='tip',data=tip_df,hue='sex')
```

```python
# Bar plot for smokers on perticular day
smoker_data = tip_df.loc[tip_df['smoker']=='Yes']

sns.barplot(x='day',y='total_bill',data=smoker_data)
```

```python
# Total bill on the basis on days and time
sns.barplot(x='time',y='total_bill',data=tip_df,hue='day')
```

```python
# bar plot for size on a perticular day
sns.barplot(x='day',y='size',data=tip_df,hue='sex')
```

```python
sns.histplot(tip_df['total_bill'],kde=True,color='orange')
```

```python
# Hist plot for man's total bill
men_bill = tip_df.loc[tip_df['sex']=='Male']
sns.histplot(men_bill['total_bill'],color='navy',kde=True)
```

```python
# Hist plot for feman's total bill
femen_bill = tip_df.loc[tip_df['sex']=='Female']
sns.histplot(femen_bill['total_bill'],color='deeppink',kde=True)
```

```python
# Hist plot for smokers total bill
smoker_bill = tip_df.loc[tip_df['smoker']=='Yes']
sns.histplot(smoker_bill['total_bill'],color='gray',kde=True)
```

```python
# Histplot of size
sns.histplot(tip_df['size'])
```

```python
# pie plot for bill in days
sun_bill = tip_df.loc[tip_df['day']=='Sun','total_bill'].sum()

sat_bill = tip_df.loc[tip_df['day']=='Sat','total_bill'].sum()

fri_bill = tip_df.loc[tip_df['day']=='Fri','total_bill'].sum()

thur_bill = tip_df.loc[tip_df['day']=='Thur','total_bill'].sum()


plt.pie([sun_bill,sat_bill,fri_bill,thur_bill],
        labels=['sun_bill','sat_bill','fri_bill','thur_bill'],
        autopct="%0.1f%%",
        shadow=True,
        explode=[0.05,0.05,0.05,0.05],
        wedgeprops={'linewidth':2,'edgecolor':'k'}
       )
plt.show()
```

```python
# pie plot for bill by male and female
male_bill = tip_df.loc[tip_df['sex']=='Male','total_bill'].sum()

female_bill = tip_df.loc[tip_df['sex']=='Female','total_bill'].sum()


plt.pie([male_bill,female_bill],
        labels=['male_bill','female_bill'],
        autopct="%0.1f%%",
        shadow=True,
        explode=[0.05,0.05],
        wedgeprops={'linewidth':2,'edgecolor':'k'}
       )
plt.show()
```
